```python
# Initialization
import glob, os, shutil
from PIL import Image, ImageDraw, ImageFont
```

```python
# Load File List
directory = r'C:\Users\Darren Wu\Desktop\SpaceInfos\2023\TW1Cont\MoRIC\Batch\SortedPreviews'
directory = directory.replace(os.sep, '/')
os.chdir(directory)
p = glob.glob('*.jpg')
print('File list loaded.')
# print(p)
```

```python
# Annotate on images
for filename in p:
    img = Image.open(filename)

    infos = filename.split('_')
    timestamp = infos[2].split('.')[0]
    infos[2] = timestamp[0:4] + '-' + timestamp[4:6] + '-' + timestamp[6:8] + 'T' + timestamp[8:10] + ':' + timestamp[10:12] + ':' + timestamp[12:14] + 'Z'

    I1 = ImageDraw.Draw(img)
    
    font_size = 15
    font_size_2 = 12
    font = ImageFont.truetype('msyh.ttc', size=font_size, encoding='gb')
    font_2 = ImageFont.truetype('msyh.ttc', size=font_size_2, encoding='gb')
    pos_x = 745
    pos_y = 15
    line_height = 22
    
    I1.text((pos_x, pos_y), '探测周期 / Observation Cycle: ' + infos[0], font=font, fill=(200, 25, 0))
    I1.text((pos_x, pos_y + line_height), '数据编号 / Data Number: ' + infos[1], font=font, fill=(200, 75, 0))
    I1.text((pos_x, pos_y + 2*line_height), '获取时间 / Acquisition Time: ' + infos[2], font=font, fill=(200, 100, 0))
    I1.text((pos_x, img.size[1]-font_size_2-2*pos_y-17), 'CNSA/PEC | @WLR2678', font=font_2, fill=(150, 150, 150))
    I1.text((pos_x, img.size[1]-font_size_2-2*pos_y), 'HX1-Or_GRAS_MoRIC-F-' + infos[1] + '_SCI_N_' + timestamp + '_' + timestamp + '_' + infos[0] + '-A', font=font_2, fill=(150, 150, 150))
    
    img.save(filename, format='JPEG', subsampling='keep', quality=100)
    print('Annotated ' + filename)
```
