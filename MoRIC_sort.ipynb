```python
# Initialization
import glob, os, shutil
```

```python
# Load File List
directory = r'C:\Users\Darren Wu\Desktop\SpaceInfos\2023\TW1Cont\MoRIC\Batch'
directory = directory.replace(os.sep, '/')
os.chdir(directory)
p = glob.glob('Preview_*.png')
print('File list loaded.')
# print(p)
```

```python
# Sort preview images according to orbit number and data number in a new subfolder
if 'SortedPreviews' not in os.listdir(directory):
    os.mkdir('SortedPreviews')
for filename in p:
    data_number = filename.split('_')[3][-4:]
    orbit_number = filename.split('_')[-2]
    timestamp = filename.split('_')[6]
    print('Data Number: ' + data_number + '; Orbit Number: ' + orbit_number + '; Timestamp: ' + timestamp)
    shutil.copy(filename, 'SortedPreviews\\' + orbit_number + '_' + data_number + '_' + timestamp + '.png')
```
